```python
import numpy as np
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision
```

```python
class MediaPipeProcessor:
    def __init__(self, model_path="./hand_landmarker.task"):
        base_options = python.BaseOptions(model_path)
        options = vision.HandLandmarkerOptions(base_options=base_options, num_hands=2)
        self.landmarker = vision.HandLandmarker.create_from_options(options)

    def processing(self, image_path):
        image = mp.Image.create_from_file(image_path)
        landmarks = self.landmarker.detect(image)

        if (landmarks.hand_landmarks):
            keypoints = []

            for hand in landmarks.hand_landmarks:
                for lm in hand:
                    keypoints.append([lm.x, lm.y, lm.z])

            keypoints = np.array(keypoints).flatten()

            return keypoints
        
        return None
```

```python
processor = MediaPipeProcessor()
processor.processing(image_path="./test_hand.jpg")
```
